#### (SPEED) Loss computation

```python
import numpy as np
import timeit

def RMSE(cal, t):
    return float(np.linalg.norm(cal - t, 1) ** 2)
def poisson_deviance(cal, t):
    eps = 1e-10
    cal = np.asarray(cal).astype(float)
    prediction_nonpositive_mask = cal < eps
    prediction_positive_mask = ~prediction_nonpositive_mask

    t_np = t[prediction_nonpositive_mask]
    pred_np = cal[prediction_nonpositive_mask]

    t_p = t[prediction_positive_mask]
    pred_p = cal[prediction_positive_mask]

    # Continuous Piecewise Loss function, to inform gradient even if outside of the domain
    loss_np = np.sum(t_np * np.log((t_np + eps) / eps) - (t_np + pred_np - 2 * eps))
    loss_p = np.sum(t_p * np.log((t_p + eps) / pred_p) - (t_p - pred_p))
    total_loss = (loss_np + loss_p)

    return total_loss
def cheap_poisson(cal, t):
    eps = 1e-6
    cal = np.where(cal > 0, cal, eps)
    return np.sum(t * np.log((t + eps) / cal) - (t - cal))
def robust_loss(cal, t):
    elem_loss = lambda y_pred, y_true: (y_pred - y_true)*np.tanh((y_pred-y_true)/2)
    return float(np.sum(elem_loss(cal, t)))

t = df.y.values
cal = t + np.sqrt(t) * 5 * (2 * np.random.rand(80) - 1)
for func in [RMSE, robust_loss, poisson_deviance, cheap_poisson]:
    name = func.__name__
    time = timeit.timeit(lambda : func(cal, t), number=100_000)
    loss = func(cal, t)
    print(f'NAME: {name}, TIME: {time:.4f}s, LOSS: {loss}')
```

```python
import numpy as np
import ast
import timeit

# 1. Extended AST-based function generator with support for exp, log, sin, cos, sqrt, etc.
def make_func_from_expr_ast_ext(expr):
    expr_ast = ast.parse(expr, mode='eval').body

    class ReplaceNames(ast.NodeTransformer):
        def visit_Name(self, node):
            # Map variable names X1, X2, ... to X[index]
            if node.id.startswith('X'):
                idx = int(node.id[1:]) - 1
                return ast.Subscript(
                    value=ast.Name(id='X', ctx=ast.Load()),
                    slice=ast.Index(ast.Constant(idx)), ctx=node.ctx)
            # Map parameter names C1, C2, ... to c[index]
            if node.id.startswith('C'):
                idx = int(node.id[1:]) - 1
                return ast.Subscript(
                    value=ast.Name(id='c', ctx=ast.Load()),
                    slice=ast.Index(ast.Constant(idx)), ctx=node.ctx)
            # Map function names to numpy.<name>
            math_funcs = {'exp', 'log', 'sin', 'cos', 'sqrt'}
            if node.id in math_funcs:
                return ast.Attribute(value=ast.Name(id='np', ctx=ast.Load()),
                                     attr=node.id, ctx=node.ctx)
            return node

    transformer = ReplaceNames()
    new_body = transformer.visit(expr_ast)
    ast.fix_missing_locations(new_body)

    # Build function AST
    func_def = ast.FunctionDef(
        name='f_ext',
        args=ast.arguments(
            posonlyargs=[], 
            args=[ast.arg(arg='X', annotation=None), ast.arg(arg='c', annotation=None)],
            vararg=None, kwonlyargs=[], kw_defaults=[], kwarg=None, defaults=[]),
        body=[ast.Return(new_body)],
        decorator_list=[]
    )
    module = ast.Module(body=[func_def], type_ignores=[])
    ast.fix_missing_locations(module)

    # Compile and execute with numpy in namespace
    env = {'np': np}
    exec(compile(module, '<ast_ext>', 'exec'), env, env)
    return env['f_ext']

# Test setup
np.random.seed(1)
X = np.random.rand(4, 50)
expr = 'exp(X1) + log(C1 * X2) - sqrt(X3) * sin(C2 * X4)'
c = np.array([2.3, 0.5])

# Instantiate extended AST function
ast_func_ext = make_func_from_expr_ast_ext(expr)

# Compute AST output
out_ast = ast_func_ext(X, c)

# Compute expected output with numpy directly
out_expected = np.exp(X[0]) + np.log(c[0] * X[1]) - np.sqrt(X[2]) * np.sin(c[1] * X[3])

# Verify correctness
max_error = np.max(np.abs(out_ast - out_expected))

# Timing
runs = 1000
time_ast_ext = timeit.timeit(lambda: ast_func_ext(X, c), number=runs)
print(time_ast_ext)
```

```python
import pandas as pd
import numpy as np
import re

def dataset_builder(
    df: pd.DataFrame,
    formula: str,
    normalization: str = 'none'
) -> tuple[list[np.ndarray], list[np.ndarray]]:
    if '~' not in formula or formula.count('~') != 1:
        raise ValueError("Formula must contain exactly one '~', e.g., 'y ~ X + C(Z)'")
    lhs, rhs = [s.strip() for s in formula.split('~')]
    if not lhs:
        raise ValueError("Left-hand side (response) is empty in formula.")
    if not rhs:
        raise ValueError("Right-hand side (predictors) is empty in formula.")
    valid_norms = {'none', 'zscore', 'minmax', 'gmd'}
    if normalization not in valid_norms:
        raise ValueError(f"Unknown normalization '{normalization}'. Choose from {valid_norms}.")

    rhs_terms = [t.strip() for t in rhs.split('+')]
    if any(not t for t in rhs_terms):
        raise ValueError("Empty term detected in RHS; check for consecutive '+' or trailing '+'.")
    cat_cols = []
    for t in rhs_terms:
        m = re.fullmatch(r'C\((\w+)\)', t)
        if m:
            col = m.group(1)
            if col not in df.columns:
                raise KeyError(f"Grouping column '{col}' not found in DataFrame.")
            cat_cols.append(col)
    feature_terms = [t for t in rhs_terms if not (re.fullmatch(r'C\((\w+)\)', t) and re.fullmatch(r'C\((\w+)\)', t).group(1) in cat_cols)]

    def parse_term(term: str, df_sub: pd.DataFrame) -> np.ndarray:
        m = re.fullmatch(r'C\((\w+)\)', term)
        if m:
            col = m.group(1)
            return pd.get_dummies(df_sub[col].astype('category'), prefix=col, drop_first=True).values
        m = re.fullmatch(r'log\((\w+)\)', term)
        if m:
            col = m.group(1); arr = df_sub[col].to_numpy(dtype=float)
            if np.any(arr <= 0): raise ValueError(f"log() requires all positive values in '{col}'.")
            return np.log(arr)
        m = re.fullmatch(r'sqrt\((\w+)\)', term)
        if m:
            col = m.group(1); arr = df_sub[col].to_numpy(dtype=float)
            if np.any(arr < 0): raise ValueError(f"sqrt() requires non-negative values in '{col}'.")
            return np.sqrt(arr)
        m = re.fullmatch(r'(\w+)\*\*(\d+)', term)
        if m:
            col,p = m.groups()
            return df_sub[col].to_numpy(dtype=float) ** int(p)
        if '*' in term and '**' not in term:
            parts = []
            for sub in term.split('*'):
                arr = parse_term(sub, df_sub)
                parts.append(arr.reshape(-1,1) if arr.ndim==1 else arr)
            res = parts[0]
            for p in parts[1:]: res = res * p
            return res
        if term in df_sub.columns:
            return df_sub[term].to_numpy()
        raise ValueError(f"Unrecognized term '{term}'. Check your formula syntax.")

    def normalize(X: np.ndarray) -> np.ndarray:
        if normalization == 'none': return X
        if normalization == 'zscore': return (X - X.mean(0)) / X.std(0)
        if normalization == 'minmax': return (X - X.min(0)) / (X.max(0) - X.min(0))
        if normalization == 'gmd':
            def gmd(arr):
                sx = np.sort(arr); n=len(arr)
                return 2 * np.sum(np.abs(np.subtract.outer(sx,sx))) / (n**2)
            scale = np.apply_along_axis(gmd, 0, X)
            return X / scale

    groups = df.groupby(cat_cols, sort=False) if cat_cols else [((), df)]
    X_list, y_list = [], []
    for _, grp in groups:
        parts = []
        for term in feature_terms:
            arr = parse_term(term, grp)
            parts.append(arr.reshape(-1,1) if arr.ndim==1 else arr)
        X = np.hstack(parts) if parts else np.empty((len(grp),0))
        X = normalize(X)
        y = parse_term(lhs, grp)
        if y.ndim != 1:
            raise ValueError("Response must be a single vector.")
        X_list.append(X); y_list.append(y)
    return X_list, y_list

def run_tests():
    # Grouping
    df1 = pd.DataFrame({'X':[1,2,3],'Z':['a','b','a'],'y':[0,1,0]})
    Xs, ys = dataset_builder(df1, 'y ~ X + C(Z)', 'none')
    np.testing.assert_array_equal(Xs[0].flatten(), [1,3])
    np.testing.assert_array_equal(ys[1], [1])

    # Minmax
    df2 = pd.DataFrame({'X':[0,5,10],'y':[1,2,3]})
    Xs2, _ = dataset_builder(df2, 'y ~ X', 'minmax')
    np.testing.assert_allclose(Xs2[0].flatten(), [0,0.5,1.0])

    # Error tests
    try: dataset_builder(df1, 'y X + C(Z)')
    except ValueError as e: assert "exactly one '~'" in str(e)
    try: dataset_builder(df1, ' ~ X')
    except ValueError as e: assert "Left-hand side" in str(e)
    try: dataset_builder(df1, 'y ~ ')
    except ValueError as e: assert "Right-hand side" in str(e)
    try: dataset_builder(df1, 'y ~ unknown', 'none')
    except ValueError as e: assert "Unrecognized term 'unknown'" in str(e)
    try: dataset_builder(df1, 'y ~ X', 'bad')
    except ValueError as e: assert "Unknown normalization" in str(e)
    print("All error-handling tests passed.")

run_tests()
```

```python
import re

# Define patterns
_MATH_FUNCS = {'exp', 'log'}
_pattern_funcs = re.compile(
    r"(?:"
    + "|".join(
        rf"{fn}\(\s*C\s*\)"   # for each fn, match fn(   C   )
        for fn in sorted(_MATH_FUNCS)
    )
    + r")"
)
_pattern_mul   = re.compile(r"\b[0-9]+(?:\.[0-9]+)?\*C\b|C\*[0-9]+(?:\.[0-9]+)?|C\*C|C\*\*[0-9]+(?:\.[0-9]+)?")  
_pattern_dup = re.compile(r'C\s*\+\s*C(?![\*/])')

def prune_poly_c_fast(eq: str) -> str:
    s = eq
    prev = None
    while s != prev:
        prev = s
        # 1) Collapse functions
        s = _pattern_funcs.sub("C", s)
        # 2) Collapse numeric multipliers and C*C
        s = _pattern_mul.sub("C", s)
        # 3) Collapse duplicate C + C until stable
        s = _pattern_dup.sub("C", s)
    return s


def prune_poly_c(eq: str) -> str:
    """
    Reducing multiple parameters in a parameterized expression to a single parameter
    :param eq: expression string
    :return: the modified expression
    >>> prune_poly_c('C*C+C+X1*C')->"C+C*X1"
    """
    for i in range(5):
        eq_l = eq
        c_poly = ['C**' + str(i) + ".5" for i in range(1, 4)]
        c_poly += ['C**' + str(i) + ".25" for i in range(1, 4)]
        c_poly += ['C**' + str(i) for i in range(1, 4)]
        c_poly += [str(i) + "*C" for i in range(1, 4)]
        c_poly += ["C*" + str(i) for i in range(1, 4)]
        for c in c_poly:
            if c in eq:
                eq = eq.replace(c, 'C')
        eq = eq.replace('C*C', 'C')
        for func in ['exp', 'log']:
            eq = eq.replace(f'{func}(C)', 'C')
        eq = str(sp.sympify(eq))
        if eq == eq_l:
            break
    return eq

equation = 'C*2 + 1.5*C/exp(C*X1) + C**0.25/(X1 + C*exp(C + C)*C + C*1)'
# equation = 'C*0.5*log(X)'
# print(timeit.timeit(lambda: prune_poly_c(equation), number=1_000))
print(timeit.timeit(lambda: prune_poly_c_fast(equation), number=100_000))
# print(prune_poly_c(equation)) # crashes
print(prune_poly_c_fast(equation))
```

```python
import warnings
def _get_numexpr_model(symbols: str, n_features: int, n_params: int):
    """
    Turn an infix string like "C1*X1 + exp(C2*X2)" into a real
    def model(X, c): ...  that calls anp.exp etc.
    """

    return ne.NumExpr(symbols)


def cal_expression_single(model, 
                          params: list) -> tuple[np.ndarray,float]:
    """
    Evaluate `symbols` on X, c via an Autograd-compiled model,
    then compute loss_fn(pred, t).  Returns (pred, loss).
    """
    try:
        # run model (suppress numpy warnings)
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            pred = model.run(*params)
    except OverflowError:
        return None, 1.23e9
    except ValueError:
        return None, 2.34e9
    except NameError:
        return None, 3.45e9
    except ArithmeticError:
        return None, 4.56e9

    return pred


def make_predictor(model: ne.NumExpr, x):
    inames = model.input_names
    # identify which positions in the arg‐tuple are X's vs C's
    x_slots = []
    c_slots = []
    for pos, name in enumerate(inames):
        if name.startswith("X"):
            idx = int(name[1:]) - 1
            x_slots.append((pos, x[idx]))
        elif name.startswith("C"):
            c_slots.append(pos)
        else: 
            raise ValueError("Unknown variable name")
    # build a single list with X's prefilled and dummy zeros for C's
    args_list = [None] * len(inames)
    for pos, x_arr in x_slots:
        args_list[pos] = x_arr
    for pos in c_slots:
        args_list[pos] = 0.0   
    def predictor(c_vec):
        # overwrite only the C slots
        for j, pos in enumerate(c_slots):
            args_list[pos] = c_vec[j]
        return cal_expression_single(model, args_list)
    return predictor
symbols = "C1 + C2*X1"
c_len = symbols.count("C")
n_features = X.shape[0]
model = _get_numexpr_model(symbols, n_features, c_len)
X = np.arange(1, 101).reshape(2,50)
mach = make_predictor(model, X)
mach(np.array([0, 1]))
```
