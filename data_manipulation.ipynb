```python
import timeit
import re

# Setup
text = "Hello World 1.5! " * 10  # Create a text string
replace_target = "1.5"
replace_with = "Universe"
pattern = re.compile(r'\d+(\.\d+)?')

# Using replace
def test_replace():
    return text.replace(replace_target, replace_with)

# Using re.sub
def test_re_sub():
    return pattern.sub(replace_with, text)

# Benchmarking
replace_time = timeit.timeit(test_replace, number=1_000_000)
re_sub_time = timeit.timeit(test_re_sub, number=1_000_000)

print(f"replace method time: {replace_time:.6f} seconds")
print(f"re.sub method time: {re_sub_time:.6f} seconds")
```

#### (SPEED) Loss computation

```python
import numpy as np
import timeit

def RMSE(cal, t):
    return float(np.linalg.norm(cal - t, 1) ** 2)
def poisson_deviance(cal, t):
    eps = 1e-10
    cal = np.asarray(cal).astype(float)
    prediction_nonpositive_mask = cal < eps
    prediction_positive_mask = ~prediction_nonpositive_mask

    t_np = t[prediction_nonpositive_mask]
    pred_np = cal[prediction_nonpositive_mask]

    t_p = t[prediction_positive_mask]
    pred_p = cal[prediction_positive_mask]

    # Continuous Piecewise Loss function, to inform gradient even if outside of the domain
    loss_np = np.sum(t_np * np.log((t_np + eps) / eps) - (t_np + pred_np - 2 * eps))
    loss_p = np.sum(t_p * np.log((t_p + eps) / pred_p) - (t_p - pred_p))
    total_loss = (loss_np + loss_p)

    return total_loss
def cheap_poisson(cal, t):
    eps = 1e-6
    cal = np.where(cal > 0, cal, eps)
    return np.sum(t * np.log((t + eps) / cal) - (t - cal))
def robust_loss(cal, t):
    elem_loss = lambda y_pred, y_true: (y_pred - y_true)*np.tanh((y_pred-y_true)/2)
    return float(np.sum(elem_loss(cal, t)))

t = df.y.values
cal = t + np.sqrt(t) * 5 * (2 * np.random.rand(80) - 1)
for func in [RMSE, robust_loss, poisson_deviance, cheap_poisson]:
    name = func.__name__
    time = timeit.timeit(lambda : func(cal, t), number=100_000)
    loss = func(cal, t)
    print(f'NAME: {name}, TIME: {time:.4f}s, LOSS: {loss}')
```

#### (SPEED) Call expression

```python
import timeit
import sympy as sp
inner_times = 100
# Arguments
X = df[['A', 'T', 'H', 'Dm']].values.T
expr = 'C1 * X1 + 5 - C2 * X3'
c = np.array([2.3, 3.1])
# Functions to benchmark
## Using eval
def cal_once_eval():
    def process_symbol_with_c(symbols, c):
        for idx, val in enumerate(c):
            symbols = symbols.replace(f"C{idx + 1}", str(val))
        return symbols
    for idx, val in enumerate(X):
        locals()[f'X{idx + 1}'] = val
    for i in range(inner_times):
        target = process_symbol_with_c(expr, c)
        eval(target)
## Using sympy
def cal_once_sympy():
    num_variables = X.shape[1]
    var_columns = [X[:, i] for i in range(X.shape[1])]
    var_symbols = sp.symbols([f'X{idx}' for idx in range(1, num_variables+1)])
    c_symbols = sp.symbols([f'C{idx}' for idx in range(1, len(c)+1)])
    args = var_symbols + c_symbols
    lambdified_func = sp.lambdify(args, expr, 'math')
    def compute_loss(params_vals):
            return lambdified_func(*var_columns, *params_vals)
    for i in range(inner_times):
         compute_loss(c)
## Using pre-compile
def cal_once_precompile():
    expr_code = expr
    num_variables = X.shape[1]
    num_c = len(c)
    for i in range(num_variables):
        expr_code = expr_code.replace(f'X{i+1}', f'X[{i}]')
    for i in range(num_c):
        expr_code = expr_code.replace(f'C{i+1}', f'C[{i}]')
    _code = compile(expr_code, '<expr>', 'eval')
    for i in range(inner_times):
        eval(_code, {'C': c, 'X': X, 'np': np})
## Using exec (TODO)
def cal_once_exec():
    # src = 
    expr_code = expr
    num_variables = X.shape[1]
    num_c = len(c)
    for i in range(num_variables):
        expr_code = expr_code.replace(f'X{i+1}', f'X[{i}]')
    for i in range(num_c):
        expr_code = expr_code.replace(f'C{i+1}', f'C[{i}]')
    _code = compile(expr_code, '<expr>', 'eval')
    for i in range(inner_times):
        eval(_code, {'C': c, 'X': X, 'np': np})
# Results
print(timeit.timeit(cal_once_eval, number=100))
print(timeit.timeit(cal_once_sympy, number=100))
print(timeit.timeit(cal_once_precompile, number=100))
```
